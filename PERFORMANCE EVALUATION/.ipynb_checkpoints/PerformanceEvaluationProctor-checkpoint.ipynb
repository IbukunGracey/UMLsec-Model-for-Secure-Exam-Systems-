### Proctor Analysis

```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
proctor = pd.read_excel("Performance Evaluation Form (Proctor) (Responses).xlsx")
```

```python
proctor.head()
```

```python
#copying the proctor dataset into another df
proctorC = proctor.copy()
```

```python
proctorC.drop_duplicates(subset="Full name", inplace= True)
```

```python
proctorC.tail()
```

```python
proctorC_info = proctorC.iloc[:, 0:6] #seleceting all rows and the first 6 columns
proctorC_info.head()
```

```python
proctorC_percept = proctorC.iloc[:, 6:20] # selecting all rows and the 7th to 21st column
proctorC_percept.head()
```

```python
proctorC_percept.columns = ["PU1_Efficient", "PU2_Convinence", "PU3_DifficultyToCheat", "PU4_SavesTime", "PEOU1_Comfortability",
                         "PEOU2_UserFriendly", "PEOU3_limitedChallenges", "PEOU4_AssetToEExams", "ATU1_WiseIdea", "ATU2_GoodInnovation",
                         "ATU3_PreferenceToOthers", "ATU4_RecommendToOthers", "BI1_LikeIdeaOfUsing","BI2_IntendToUse"]
```

```python
proctorC_percept.head()
```

```python
response_map= {"Strongly agree (SA)": 5, "Agree (A)": 4, "Undecided (UD)": 3, "Disagree (D)": 2,
              "Strongly Disagree (SD)": 1}
```

```python
# proctorC_percept[["PU1_Efficient", "PU2_Convinence", "PU3_DifficultyToCheat", "PU4_SavesTime", "PEOU1_Comfortability",
#                         "PEOU2_UserFriendly", "PEOU3_limitedChallenges", "PEOU4_AssetToEExams", "ATU1_WiseIdea", "ATU2_GoodInnovation",
#                         "ATU3_PreferenceToOthers", "ATU4_RecommendToOthers", "BI1_LikeIdeaOfUsing","BI2_IntendToUse"]].map(response_map)
```

```python
#converting the text to numbers
proctorC_percept["PU1_Efficient"] = proctorC_percept["PU1_Efficient"].map(response_map)
```

```python
proctorC_percept["PU2_Convinence"] = proctorC_percept["PU2_Convinence"].map(response_map)
proctorC_percept["PU3_DifficultyToCheat"] = proctorC_percept["PU3_DifficultyToCheat"].map(response_map)
proctorC_percept["PU4_SavesTime"] = proctorC_percept["PU4_SavesTime"].map(response_map)
proctorC_percept["PEOU1_Comfortability"] = proctorC_percept["PEOU1_Comfortability"].map(response_map)
proctorC_percept["PEOU2_UserFriendly"] = proctorC_percept["PEOU2_UserFriendly"].map(response_map)
proctorC_percept["PEOU3_limitedChallenges"] = proctorC_percept["PEOU3_limitedChallenges"].map(response_map)
proctorC_percept["PEOU4_AssetToEExams"] = proctorC_percept["PEOU4_AssetToEExams"].map(response_map)
proctorC_percept["ATU1_WiseIdea"] = proctorC_percept["ATU1_WiseIdea"].map(response_map)
proctorC_percept["ATU2_GoodInnovation"] = proctorC_percept["ATU2_GoodInnovation"].map(response_map)
proctorC_percept["ATU3_PreferenceToOthers"] = proctorC_percept["ATU3_PreferenceToOthers"].map(response_map)
proctorC_percept["ATU4_RecommendToOthers"] = proctorC_percept["ATU4_RecommendToOthers"].map(response_map)
proctorC_percept["BI1_LikeIdeaOfUsing"] = proctorC_percept["BI1_LikeIdeaOfUsing"].map(response_map)
proctorC_percept["BI2_IntendToUse"] = proctorC_percept["BI2_IntendToUse"].map(response_map)
```

```python
proctorC_percept
```

```python
proctorC_percept.to_csv("storedProcPer", index=False)
```
